```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
from utils import get_image_paths, word_from_image_path
```

```python
base_dir = '/mnt/mjsynth/mnt/ramdisk/max/90kDICT32px/'
```

```python
images_paths = get_image_paths(base_dir)
```

```python
%%time
words = [word_from_image_path(p).lower() for p in images_paths[:]]
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
cv = CountVectorizer(analyzer='char', ngram_range=(1, 4), max_features=10000, tokenizer=TreebankWordTokenizer().tokenize)
```

```python
%%time
cv.fit(words[:])
```

```python
from itertools import product
```

```python
for a in cv.vocabulary_:
    if '0' in a:
        print(a)
```

```python
letters = '0123456789abcdefghijklmnopqrstuvwyz'
```

```python
res = 0
for l1, l2, l3 in product(letters, letters, letters):
    if l1+l2+l3 in cv.vocabulary_:
        print(l1+l2+l3)
        res += 1
```
